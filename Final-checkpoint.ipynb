In stock market, opening bells does not actually mean that the company is making a lot of profit. With an opening bell, a stock might actually go down. Below i will take a look at how google and apple share  has either gained or tanked. My focus will be the first few months and the last few months.

```python
# Assign `Adj Close` to `daily_close`
daily_close = aapl[['Adj Close']]

daily_pct_c = daily_close.pct_change()
daily_pct_c.fillna(0, inplace=True)

#print(daily_pct_c)

# Daily log returns
daily_log_returns = np.log(daily_close.pct_change()+1)

#print(daily_log_returns)

daily_pct_c = daily_close / daily_close.shift(1) - 1
print(daily_pct_c)
```

```python
# Import matplotlib
import matplotlib.pyplot as plt

# Plot the distribution of `daily_pct_c`
daily_pct_c.hist(bins=50)
plt.ylabel('ADJ Stock Price in USD')
plt.label("Adjusted closing price index for Apple")
# Show the plot
plt.show()

# Pull up summary statistics
print(daily_pct_c.describe())
```

```python
# Assign `Adj Close` to `daily_close`
daily_closeGoogle = google[['Adj Close']]

dailyGoogle = daily_closeGoogle.pct_change()
dailyGoogle.fillna(0, inplace=True)

#print(dailyGoogle)

# Daily log returns
daily_log_returns = np.log(daily_closeGoogle.pct_change()+1)

#print(daily_log_returns)

dailyGoogle = daily_closeGoogle / daily_closeGoogle.shift(1) - 1
print(dailyGoogle)
```

```python
# Import matplotlib
import matplotlib.pyplot as plt

#Show distribution for Google
dailyGoogle.hist(bins=50)

# Show the plot
plt.show()
plt.ylabel('ADJ Stock Price in USD')
plt.label("Adjusted closing price for Google")
# Pull up summary statistics
print(dailyGoogle.describe())
```

I will be checking the cumulative daily return.

```python
# Calculate the cumulative daily returns for Apple
cum_daily_return = (1 + daily_pct_c).cumprod()

# Print `cum_daily_return`
print(cum_daily_return)
```

```python
# Calculate the cumulative daily returns for Google
cum_daily_return_google = (1 + dailyGoogle).cumprod()

# Print `cum_daily_return`
print(cum_daily_return_google)
```

```python
# Import matplotlib
import matplotlib.pyplot as plt 

# Plot the cumulative daily returns
cum_daily_return['Adj Close'].plot()
cum_daily_return_google['Adj Close'].plot()

# Show the plot
plt.show()
```
